```python
# =========================================
# 1. Import libraries
# =========================================
import pandas as pd
import numpy as np

# =========================================
# 2. Load the dataset
# =========================================
url = "https://raw.githubusercontent.com/ywchiu/riii/master/data/house-prices.csv"
df = pd.read_csv(url)

# =========================================
# 3. Initial exploration of the dataset
# =========================================

print("First 5 rows of the dataset:")
display(df.head())

print("\nDataset shape (rows, columns):")
print(df.shape)

print("\nList of columns:")
print(df.columns)

print("\nData types:")
print(df.dtypes)

print("\nSummary statistics:")
display(df.describe())

# Confirm the target variable
print("\nTarget variable:")
print("Price")

# =========================================
# 4. Identify uninformative columns
# =========================================
print("\nChecking possible ID-like columns:")
print(df.head())

# 'Home' is just an identifier for each house
# so it is not informative for prediction.

# =========================================
# 5. Data preprocessing
# =========================================

# ---- Remove unnecessary column
df = df.drop(columns=["Home"])

print("\nDataset after removing uninformative columns:")
display(df.head())

# ---- Check missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# If missing values existed, we could handle them with:
# df = df.dropna()
# or
# df.fillna(df.mean())

# In this dataset normally there are no missing values.

# ---- Verify data types
print("\nData types after cleaning:")
print(df.dtypes)

# Convert categorical variables to numeric
df = pd.get_dummies(df, columns=["Brick", "Neighborhood"], drop_first=True)

print("\nDataset after encoding categorical variables:")
display(df.head())

# =========================================
# 6. Define X and y for modeling
# =========================================

y = df["Price"]          # Target variable
X = df.drop(columns=["Price"])  # Input features

print("\nShape of X:", X.shape)
print("Shape of y:", y.shape)

print("\nFeatures used for modeling:")
print(X.columns)

print("\nTarget variable preview:")
print(y.head())
```

4

```python
# =========================================
# 4. Split the Dataset
# =========================================

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,   # 20% testing
    random_state=42
)

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

5

```python
# =========================================
# 5. Feature Scaling
# =========================================

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit only on training data
X_train_scaled = scaler.fit_transform(X_train)

# Apply same transformation to test data
X_test_scaled = scaler.transform(X_test)

print("Feature scaling completed.")
```

6

```python
# =========================================
# 6. Create the Regression Model
# =========================================

from sklearn.linear_model import SGDRegressor

sgd_model = SGDRegressor(
    max_iter=1000,
    learning_rate='constant',
    eta0=0.01,
    random_state=42
)

print(sgd_model)
```

7

```python
# =========================================
# 7. Train the Model
# =========================================

sgd_model.fit(X_train_scaled, y_train)

print("Model training completed.")
```

8

```python
# =========================================
# 8. Generate Predictions
# =========================================

y_pred = sgd_model.predict(X_test_scaled)

print("First predictions:")
print(y_pred[:5])
```

9

```python
# =========================================
# 9. Model Evaluation
# =========================================

from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
```
